# standardize text values

```python
data["country"]=data["country"].str.lower().str.strip()
data["type"]=data["type"].str.lower().str.strip()
data["title"]=data["title"].str.lower().str.strip()
data["director"]=data["director"].str.lower().str.strip()
data["cast"]=data["cast"].str.lower().str.strip()
data["listed_in"]=data["listed_in"].str.lower().str.strip()
```

# changing NaN to ' ' :-

```python
data["director"]=data["director"].fillna('').astype(str)
data["cast"]=data["cast"].fillna('').astype(str)
data["country"]=data["country"].fillna('').astype(str)
data["date_added"]=data["date_added"].fillna('').astype(str)
data["rating"]=data["rating"].fillna('').astype(str)
```

```python
data["type"]=data["type"].apply(str)
data["title"]=data["title"].apply(str)
data["director"]=data["director"].apply(str)
data["cast"]=data["cast"].apply(str)
data["country"]=data["country"].apply(str)
data["date_added"]=data["date_added"].apply(str)
data["rating"]=data["rating"].apply(str)
data["duration"]=data["duration"].apply(str)
data["listed_in"]=data["listed_in"].apply(str)
```
